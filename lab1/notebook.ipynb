```python
# 1.
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from typing import Union, Tuple
```

```python
Number = Union[int, float]

def genLabel(i: int) -> str:
    return chr(ord('a') + i)

def genRandDist(fname: str, size: Union[Tuple, Number], mu: Number = 0, sigma = 1) -> np.ndarray:
    return mu + sigma * getattr(np.random, fname)(size=size)

# 2.
distributions = [
    # a) 1000 samples of uniform distribution on interval 0-1
    [genRandDist(fname="uniform", size=1000), "uniform 1000"],

    # b) 1000 samples of normal distribution with mean 0 and standard deviation 1
    [genRandDist(fname="standard_normal", size=1000, sigma=1), "st. norm. 1000, std=1"],

    # c) 1000 samples of normal distribution with mean 0 and standard deviation 10
    [genRandDist(fname="standard_normal", size=1000, sigma=10), "st. norm. 1000, std=10"],

    # d - f 1000 samples of beta distribution for the following pairs of parameters a, b in [(1,1), (2,2), (3,9)]
    *[[np.random.beta(a=a, b=b,size=1000), f"beta 1000, a={a}, b={b}"] for a, b in [(1,1), (2,2), (3,9)]],

    # g - h 10000 samples of two-dimensional normal distributions each with mean [0, 0].T and covariance matrices
    # S1 = [[1, 0], [0, 1]], S2 = [[2, 1], [1, 2]]
    # note: if we wanted to get 10000 in total (i.e., not just in the first dimension), then we would do size=5000,
    # so that the dims (5000, 2) implicated a total of 10000 samples; here, however, we want the output to be 10000 x 2
    # g) for S1
    [np.random.multivariate_normal(size=10000, mean=np.array([0, 0]).T, cov=[[1, 0], [0, 1]]), "norm 2D 10000, S1"],
    # h) for S2
    [np.random.multivariate_normal(size=10000, mean=np.array([0, 0]).T, cov=[[2, 1], [1, 2]]), "norm 2D 10000, S1"],
]

for i, (dist, title) in enumerate(distributions):
    print(f"{genLabel(i)}) {title}")
    print(f"\tCalculated real params: mu={np.mean(dist) :.3f}, std={np.std(dist) :.3f}, shape={dist.shape}")
    display(dist)
    print("") # new line
```

```python
# 3.

distributionsOnly1D = [pack for pack in distributions if len(pack[0].shape) == 1]
ct = 0

# a) uniform 1000
(dist, title) = distributionsOnly1D[ct]

plt.figure(figsize=(8, 6))

_, bins, _ = plt.hist(x=dist, bins="auto", density=True)
plt.xlabel("X")
plt.ylabel("Count")
plt.title(f"2D Histogram for {genLabel(ct)}) {title}")
plt.grid()

plt.plot(bins, np.ones_like(bins))

plt.legend(["Calculated", "Analytical"])
```

```python
from scipy import stats

# b) st. norm. 1000, std=1

ct = 1
mu = 0
sigma = 1

(dist, title) = distributionsOnly1D[ct]

plt.figure(figsize=(8, 6))

_, bins, _ = plt.hist(x=dist, bins="auto", density=True)
plt.xlabel("X")
plt.ylabel("Count")
plt.title(f"2D Histogram for {genLabel(ct)}) {title}")
plt.grid()

x = np.linspace(mu - 5 * sigma, mu + 5 * sigma, 100)

plt.plot(x, stats.norm.pdf(x, mu, sigma))

plt.legend(["Calculated", "Analytical"])
```

### Ad 3: Comparison with analytical formulas

The plots approximately match with the shape of plots from analytical formulas.

```python
# 4.

distributionsOnly2D = [pack for pack in distributions if len(pack[0].shape) == 2]

for i, (dist, title) in enumerate(distributionsOnly2D):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
    axes = axes.ravel()

    fig.suptitle(f"2D histogram for {genLabel(i)}) {title}", fontsize=14)

    ax = axes[0]
    ax.hist2d(x=dist[:, 0], y=dist[:, 1])
    ax.set_xlabel("X")
    ax.set_ylabel("Count")
    ax.set_title("Matplotlib - hist2d")
    ax.grid()

    # # creates a separate figure, no way to override...
    # sns.jointplot(data=dist)
    # plt.title("Seaborn - jointplot")

    ax = axes[1]
    sns.histplot(x=dist[:, 0], y=dist[:, 1], ax=ax)
    ax.set_title("Seaborn - kdeplot")
    ax.grid()
```

### Ad 4
The plots from both libraries only differ in style, but not with critical terms like the scale, shape or values. However, seaborn is much more precise, while matplotlib seems to have aggregated the data over a sliding window.

```python
# 5.
size = 1000

a = genRandDist(fname="normal", size=size, mu=2, sigma=0.1)
b = genRandDist(fname="normal", size=size, mu=0.2, sigma=0.2)
x = genRandDist(fname="normal", size=size, mu=0, sigma=3)
sigma = np.abs(genRandDist(fname="normal", size=size))

mu = a + b * x
y = genRandDist(fname="normal", size=size, mu=mu, sigma=sigma)

plt.figure(figsize=(10, 8))
plt.scatter(x, y)
miLinesX = np.clip(x, -6, 6)

for i in range(size):
    plt.plot(miLinesX, a[i] + b[i] * x, "orange", alpha=0.1)

plt.title("$y(x)$")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.legend(["$y(x)$", "$y = \\alpha + \\beta x$"])
```
